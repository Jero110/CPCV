```
from cpcv_analysis.data import load_data
from cpcv_analysis.backtest_engine import (
    get_last_n_days, cpcv_debug, cpcv_sharpe_dist, compare_methods
)
from xgboost import XGBClassifier
from cpcv_analysis.config import XGB_PARAMS

X_full, y_full, t1_full, prices, fwd_ret_full = load_data()
X, y, t1, fwd_ret = get_last_n_days(X_full, y_full, t1_full, fwd_ret_full, n=100)
print(f"Datos: {len(X)} obs  {X.index[0].date()} → {X.index[-1].date()}")
```

```
clf = XGBClassifier(**XGB_PARAMS)
```

```
cpcv_sharpes = cpcv_debug(clf, X, y, t1, fwd_ret, split_to_inspect=0)
print("\nSharpes de paths (debug):")
print(cpcv_sharpes.to_string())
```

```
cpcv_prod = cpcv_sharpe_dist(clf, X, y, t1, fwd_ret)
print("Sharpes CPCV (producción):", cpcv_prod.values)
print("¿Coinciden con debug?", (cpcv_sharpes.values == cpcv_prod.values).all())
```

```
results = compare_methods(clf, X, y, t1, fwd_ret)
```
